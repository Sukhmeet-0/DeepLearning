```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('Mall_Customers.csv')
x = data.iloc[:, [3,4]].values
```

```python
from scipy.cluster import hierarchy as sch 
dendrogram = sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('dendrogram')
plt.xlabel('customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5,linkage='ward')
res=ac.fit_predict(x)
print(res)
```

```python
plt.scatter(x[res==0,0],x[res==0,1],c='red',s=100,label='cluster 1')
plt.scatter(x[res==1,0],x[res==1,1],c='green',s=100,label='cluster 2')
plt.scatter(x[res==2,0],x[res==2,1],c='blue',s=100,label='cluster 3')
plt.scatter(x[res==3,0],x[res==3,1],c='yellow',s=100,label='cluster 4')
plt.scatter(x[res==4,0],x[res==4,1],c='purple',s=100,label='cluster 5')
plt.legend()
plt.title('Clusters of customers')
plt.xlabel('Annual income($)')
plt.ylabel('Spending Score')
plt.show()
```
