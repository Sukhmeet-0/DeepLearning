```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data  = pd.read_csv('Mall_Customers.csv')
```

```python
x = data.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    km = KMeans(n_clusters=i,init='k-means++',random_state=42)
    km.fit(x)
    wcss.append(km.inertia_)
plt.plot(range(1,11),wcss,color='red')
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5,init='k-means++',random_state=42)
res=kmeans.fit_predict(x)
```

```python
plt.scatter(x[res==0,0],x[res==0,1],c='red',s=100,label='cluster 1')
plt.scatter(x[res==1,0],x[res==1,1],c='green',s=100,label='cluster 2')
plt.scatter(x[res==2,0],x[res==2,1],c='blue',s=100,label='cluster 3')
plt.scatter(x[res==3,0],x[res==3,1],c='yellow',s=100,label='cluster 4')
plt.scatter(x[res==4,0],x[res==4,1],c='purple',s=100,label='cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],label='centroids',c='black',s=100)
plt.legend()
plt.title('Clusters of customers')
plt.xlabel('Annual income($)')
plt.ylabel('Spending Score')
plt.show()
```
