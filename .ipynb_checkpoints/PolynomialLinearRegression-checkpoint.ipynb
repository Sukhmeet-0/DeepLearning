```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Position_Salaries.csv')
```

```python
x=data.iloc[:,1:-1].values
y=data.iloc[:,-1].values
```

```python
data.isnull().sum()
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf =PolynomialFeatures(degree=10)
xpoly=pf.fit_transform(x)
linreg=LinearRegression()
linreg.fit(xpoly,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lr.predict(x),color='blue')
plt.title('Truth of Bluff')
plt.xlabel('position label')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,linreg.predict(xpoly),color='blue')
plt.title('Truth of Bluff')
plt.xlabel('position label')
plt.ylabel('Salary')
plt.show()
```

```python
xgrid=np.arange(min(x),max(x),.1)
xgrid=xgrid.reshape((len(xgrid),1))
plt.scatter(x,y,color='red')
plt.plot(xgrid,linreg.predict(pf.fit_transform(xgrid)),color='blue')
plt.title('Truth of Bluff')
plt.xlabel('position label')
plt.ylabel('Salary')
plt.show()
```
